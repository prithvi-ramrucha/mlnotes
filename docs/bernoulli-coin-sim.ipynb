# **Bernoulli Coin Simulation**

Loading dependencies,

```python
# Importing,
import numpy as np
import matplotlib.pyplot as plt
```

Simulation,

```python
# Parameters
N = 100000 # <-- Number of coin flips.
p = 1/2

# Simulation,
mu = 0 
counter = 1
for i in range(1, N+1):
    uniform = np.random.uniform(low=0, high=1, size=1)
    b = 0 if uniform < p else 1
    mu += (1/counter)*(b - mu) 
    counter += 1
```

Graphing,

```python
# Parameters
N = 1000 # <-- Number of coin flips.
p = 1/2

# Simulation,
mu = 0 
counter = 1
running_mean = np.empty(shape=N)
for i in range(1, N+1):
    uniform = np.random.uniform(low=0, high=1, size=1)
    b = 0 if uniform < p else 1
    mu += (1/counter)*(b - mu) 
    running_mean[(i-1)] = mu
    counter += 1

# Graphing,
plt.plot(running_mean, color="blue")
plt.xlabel("Number of Trials $n$")
plt.axhline(y=0.5, color="red", linestyle="--")
plt.ylabel("Emperical Mean")

# Saving figure,
plt.savefig("bernoulli-coin-sim.png", dpi=300)
plt.show()
```
